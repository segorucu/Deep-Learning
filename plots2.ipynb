```python
import pandas as pd
import numpy as np
from time import time
from IPython.display import display
#import seaborn as sns
#import visuals as vs
import matplotlib.pyplot as pl
import matplotlib
#import matplotlib.patches as mpatches

# Pretty display for notebooks
%matplotlib inline
```

```python
data = pd.read_csv("data/train loss/train_accuracy.csv")
```

```python
data.head(1)
```

```python
font = {'size'   : 12}

matplotlib.rc('font', **font)
```

```python

# Create figure
fig, ax = pl.subplots(1,1,figsize = (5,5))

# max_depth
ax.plot(data["Step"],data["Visual Bert"],color="red",label="Visual Bert")
ax.plot(data["Step"],data["Vilbert"],color="blue",label="Vilbert")
ax.plot(data["Step"],data["Transformers"],color="green",label="Transformers")
ax.plot(data["Step"],data["Concat Bow"],color="yellow",label="Concat Bow")
ax.plot(data["Step"],data["CNN+LSTM"],color="purple",label="CNN+LSTM")
ax.legend()
ax.set_xlabel("Epochs")
ax.set_ylabel("Training Loss")


# Aesthetics
pl.title("Training Loss")
pl.tight_layout()
#pl.show()
pl.savefig("Training_loss.png")
    
```
